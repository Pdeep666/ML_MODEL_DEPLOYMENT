**REMOVAL OF OUTLIERS**

```python
def remove_outliers(df, column):
    upper_limit = df[column].mean() + 3 * df[column].std()
    lower_limit = df[column].mean() - 3 * df[column].std()
    new_df = df.loc[(df[column] <= upper_limit) & (df[column] >= lower_limit)]
    return new_df
```

```python
new_df=remove_outliers(df,'windspeed')
new_df=remove_outliers(new_df,'dewpoint')
new_df=remove_outliers(new_df,'cloud ')
new_df=remove_outliers(new_df,'humidity ')
```

```python
X2 = df_downsampled.drop(columns=['rainfall','maxtemp', 'temparature', 'mintemp','day'], axis=1)
y2 = df_downsampled['rainfall'].dropna()
X2 = X2.loc[y2.index]
```

```python
y2.head()
```

```python
X2.head()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
# Split into 80% training and 20% testing
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)

# Display dataset shapes
print(f"Training set: X_train={X2_train.shape}, y_train={y2_train.shape}")
print(f"Testing set: X_test={X2_test.shape}, y_test={y2_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import pickle
```

**RANDON FOREST CLASSIFIER**

```python
rf_model = RandomForestClassifier(random_state=42)

param_grid_rf = {
    "n_estimators": [50, 100, 200],
    "max_features": ["sqrt", "log2"],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
# Hypertuning using GridSearchCV
grid_search_rf = GridSearchCV(estimator=rf_model, param_grid=param_grid_rf, cv=5, n_jobs=-1, verbose=2)

grid_search_rf.fit(X2_train, y2_train)
```

```python
best_rf_model = grid_search_rf.best_estimator_

print("best parameters for Random Forest:", grid_search_rf.best_params_)
```

```python
cv_scores = cross_val_score(best_rf_model, X2_train, y2_train, cv=7)
print("Cross-validation scores:", cv_scores)
print("Mean cross-validation score:", np.mean(cv_scores))
```

```python
# test set performance
y_pred = best_rf_model.predict(X2_test)

print("Test set Accuracy:", accuracy_score(y2_test, y_pred))
print("Test set Confusion Matrix:\n", confusion_matrix(y2_test, y_pred))
print("Classification Report:\n", classification_report(y2_test, y_pred))
```

**LOGESTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression

# Define parameter grid for Logistic Regression
lr_param_grid = {
    'C': [0.1, 1, 10],
    'solver': ['liblinear', 'lbfgs']
}
```

```python
# Initialize and perform Grid Search
lr_grid_search = GridSearchCV(LogisticRegression(max_iter=1000, random_state=42), lr_param_grid, cv=5, scoring='accuracy', n_jobs=-1)
lr_grid_search.fit(X2_train, y2_train)
```

```python
# Best parameters and model
print("Best Parameters for Logistic Regression:", lr_grid_search.best_params_)
best_lr = lr_grid_search.best_estimator_
```

```python
cv_scores = cross_val_score(best_lr, X2_train, y2_train, cv=7)
print("Cross-validation scores:", cv_scores)
print("Mean cross-validation score:", np.mean(cv_scores))
```

```python
# test set performance
y_pred = best_lr.predict(X2_test)

print("Test set Accuracy:", accuracy_score(y2_test, y_pred))
print("Test set Confusion Matrix:\n", confusion_matrix(y2_test, y_pred))
print("Classification Report:\n", classification_report(y2_test, y_pred))
```
